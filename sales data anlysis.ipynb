```python
# sales by rejion
sns.barplot(x = "Region" , y = "Sales",data = df , estimator=np.sum)
plt.title("Sales by region")
plt.show()
```

```python
# profit by category
sns.barplot(x = "Category", y = "Profit" , data=df , estimator=np.sum)
plt.title("Profit by Category")
plt.show()
```

```python
# Daily sales trend
daily_sales.plot(marker = "0")
plt.title("Daily sales trend")
plt.show()
```

```python
# Correlation heatmap
sns.heatmap(
    df[["Sales","Quantity","Discount","Profit"]].corr(),
    annot = True
)
plt.title("Correlation heatmap")
plt.show()
       
```
